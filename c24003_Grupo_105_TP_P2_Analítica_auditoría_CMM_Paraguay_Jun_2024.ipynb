### **Archivo sales_in_Paraguay: Resumen de ventas en Paraguay**

### Preparado de tablas para visualizaciones

##### **- Renombrado de dataframes**

```python
df_exports.rename(columns={'Rubber':'Rubber_exp','Brass':'Brass_exp','Vinyl':'Vinyl_exp','Granite':'Granite_exp','Stone':'Stone_exp',
                         'Brick':'Brick_exp','Aluminum':'Aluminum_exp','Glass':'Glass_exp','Plexiglass':'Plexiglass_exp','Steel':'Steel_exp',
                         'Wood':'Wood_exp','Plastic':'Plastic_exp'}, inplace = True)
print(df_exports.head())
```

```python
df_sales.rename(columns={'Rubber':'Rubber_sales','Brass':'Brass_sales','Vinyl':'Vinyl_sales','Granite':'Granite_sales','Stone':'Stone_sales',
                         'Brick':'Brick_sales','Aluminum':'Aluminum_sales','Glass':'Glass_sales','Plexiglass':'Plexiglass_sales','Steel':'Steel_sales',
                         'Wood':'Wood_sales','Plastic':'Plastic_sales'}, inplace = True)
print(df_sales.head())
```

##### **- Unión de dataframes y orden de columnas**

```python
# unión de df_exports y df_sales como df_merged
df_merged = df_exports.merge(df_sales, how='inner', on='distributor')
print(df_merged.head())
```

```python
# ordenamos las columnas
df_merged = df_merged.reindex(sorted(df_merged.columns), axis=1)
df_merged = df_merged[['distributor'] + [col for col in df_merged.columns if col != 'distributor']]
display(df_merged.head())
```

#### **- Sumarizamos los valores de las columnas**

```python
Total_merged = df_merged.iloc[:, 1:].sum(axis=0).copy()
print(Total_merged)
```

###**Visualizaciones**

Luego de limpiar, revisar y preparar las tablas para el análisis, hacemos las visualizaciones para terminar de comprender y explicar los datos.

```python
c=['mediumpurple','mediumpurple','mediumorchid','mediumorchid','crimson','crimson'
  ,'peru','peru','darkkhaki','darkkhaki','cadetblue','cadetblue'
  ,'darkgray','darkgray','dimgray','dimgray','firebrick','firebrick'
  ,'rosybrown','rosybrown','plum','plum','tab:purple','tab:purple']


fig, gra_total = plt.subplots(figsize=(14,8))
gra_total.set_title('Gráfico de Exportaciones y Ventas en Paraguay')
gra_total.barh(Total_merged.index, Total_merged.values, color=c)
gra_total.bar_label(gra_total.containers[0],  fmt="$%0.0f", backgroundcolor='white')
gra_total.set_xlabel('Valores')
gra_total.set_ylabel('Materiales')
gra_total.invert_yaxis()

plt.show()
```

## **Conclusiones**

Realizamos este análisis para intentar explicar la baja tasa de retorno de inversión de la actividad de CMM en Paraguay. Al analizar los datos aportados y las gráficas generadas con la información de la compañía, se puede concluir lo siguiente:

**La baja rentabilidad es causada por el bajo rendimiento de acero y ladrillos**

  - Tanto el acero como los ladrillos tuvieron un ratio Exportaciones/ventas negativo del **99.9%**
  
  - Del total de las exportaciones de ladrillos, **solo el 6.35% se distribuyó en zonas con actividades afines**.
  
  -El **89.67%** de las exportaciones de acero se destinó a **un solo distribuidor** que, aunque se encontraba en un área de actividades afines, **vendió solo el 0.0001%** del acero exportado.

**Basándonos en estos análisis concluimos que baja tasa de retorno se debió a un fallo de estrategia de distribución por parte de los directivos de CMM.**

## **Conclusiones adicionales**

**1. Margen positivo en la mayoría de productos**

- En la mayoría de productos, se puede ver que las ventas son superiores al material importado, es decir que tiene un margen de ganancia positivo.

**2. Mayor Retorno Relativo: Vidrio (Glass)**

- En este caso particular, el margen de ganancias del Vidrio es superior al de los demás productos, esto quiere decir que es el material que tiene mejor retorno relativo sobre la inversión, es decir que es el material más eficiente en términos del negocio.

**3. Mayor Ganancia Absoluta: Plexiglass**

- Por otra parte el material con una mayor Ganancia Absoluta es el Plexiglass, es decir que es el que más aporta a las ganancias totales pues fué el que más se vendió.

- Estas dos métricas (Margen de Ganancia y Ganancia Absoluta) son importantes a la hora de diseñar estrategias de inversión y saber donde dirigir los recursos y esfuerzos a la hora de maximizar las ganancias.
